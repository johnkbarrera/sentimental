```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import pandasql as pdsql
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as dates
from matplotlib.backends.backend_pdf import PdfPages
```

## Procesando  respuestas

### Procesando KPI de velocidad de atenci贸n

```python
#Leyendo CSV
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%y')
df_respuestas = pd.read_csv('clean_respuestas.csv',parse_dates=['entrega'], date_parser=dateparse)

pysql = lambda q: pdsql.sqldf(q, globals())

#Definiendo sentencia SQL
strsql = "select Tienda, entrega, count(*) as count_total from df_respuestas group by Tienda,entrega order by Tienda,entrega ASC"
#creando dataframe
df_total_Velocidad_de_atencion = pysql(strsql)

#Definiendo sentencia SQL
strsql = "select Tienda,entrega, count(Velocidad_de_atencion) as promotor from df_respuestas where Velocidad_de_atencion>=0.8 group by Tienda,entrega order by Tienda,entrega ASC "
#creando dataframe
df_Velocidad_de_atencion_promotor = pysql(strsql)

#Juntando dataframes en uno solo 
new_df = pd.merge(df_total_Velocidad_de_atencion, df_Velocidad_de_atencion_promotor,  how='left', 
                  left_on=['Tienda', 'entrega'], right_on = ['Tienda', 'entrega'])

#Definiendo sentencia SQL
strsql = "select Tienda,entrega, count(Velocidad_de_atencion) as detractor from df_respuestas where Velocidad_de_atencion<=0.8 group by Tienda,entrega order by Tienda,entrega ASC"
#Creando dataframe
df_Velocidad_de_atencion_detractor = pysql(strsql)

#Juntando dataframes en uno solo 
df_velocidad = pd.merge(new_df, df_Velocidad_de_atencion_detractor,  how='left', 
                  left_on=['Tienda', 'entrega'], right_on = ['Tienda', 'entrega'])

#Remplazando los valores NaN por ceros
df_velocidad.fillna(0, inplace=True)

# Extractcci贸n de listas
lista_total_Velocidad_de_atencion = df_velocidad['count_total'].tolist()
list_Velocidad_de_atencion_promotor=df_velocidad['promotor'].tolist()
list_Velocidad_de_atencion_detractor = df_velocidad['detractor'].tolist()
```

### Procesando KPI de velocidad de amabilidad

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%y')
df_respuestas = pd.read_csv('clean_respuestas.csv',parse_dates=['entrega'], date_parser=dateparse)

pysql = lambda q: pdsql.sqldf(q, globals())

strsql = "select Tienda, entrega, count(*) as count_total from df_respuestas group by Tienda,entrega order by Tienda,entrega ASC"
df_total_amabilidad = pysql(strsql)

strsql = "select Tienda,entrega, count(Amabilidad) as promotor from df_respuestas where Amabilidad>=0.8 group by Tienda,entrega order by Tienda,entrega ASC "
df_amabilidad_promotor = pysql(strsql)

new_df = pd.merge(df_total_amabilidad, df_amabilidad_promotor,  how='left', 
                  left_on=['Tienda', 'entrega'], right_on = ['Tienda', 'entrega'])

strsql = "select Tienda,entrega, count(Amabilidad) as detractor from df_respuestas where Amabilidad<=0.8 group by Tienda,entrega order by Tienda,entrega ASC"
df_amabilidad_detractor = pysql(strsql)

df_amabilidad = pd.merge(new_df, df_amabilidad_detractor,  how='left', 
                  left_on=['Tienda', 'entrega'], right_on = ['Tienda', 'entrega'])

df_amabilidad.fillna(0, inplace=True)

# Extractcci贸n de listas
lista_fecha = df_amabilidad['entrega'].tolist()
lista_tienda = df_amabilidad['Tienda'].tolist()
lista_total_amabilidad = df_amabilidad['count_total'].tolist()
list_amabilidad_promotor=df_amabilidad['promotor'].tolist()
list_amabilidad_detractor = df_amabilidad['detractor'].tolist()
```

### Calculo de ntp

```python
list_ntp = []

size = len(lista_fecha)
i = 0 
while i< size:
    amabilidad_promotor = list_amabilidad_promotor[i]
    amabilidad_detractor = list_amabilidad_detractor[i]
    total_amabilidad = lista_total_amabilidad[i]
    amabilidad_ntp  = (amabilidad_promotor - amabilidad_detractor)/total_amabilidad
    
    velocidad_promotor = list_Velocidad_de_atencion_promotor[i]
    velocidad_detractor = list_Velocidad_de_atencion_detractor[i]
    total_velocidad = lista_total_Velocidad_de_atencion[i]
    velocidad_de_atencion_ntp  = (velocidad_promotor - velocidad_detractor)/total_velocidad
    
    list_ntp.append( (velocidad_de_atencion_ntp + amabilidad_ntp)/2 )
    i = i + 1

se = pd.Series(list_ntp)
df_amabilidad['ntp']=se
df_ntp = df_amabilidad
df_ntp.drop('count_total', axis=1, inplace=True)
df_ntp.drop('promotor', axis=1, inplace=True)
df_ntp.drop('detractor', axis=1, inplace=True)
df_ntp['entrega'] = pd.to_datetime(df_ntp['entrega'])
p = df_ntp.pivot(index='entrega', columns='Tienda', values='ntp')
p.fillna(0, inplace=True)
df_resumen = p.describe()
p.to_csv("consolidado_tiendas_ntp.csv", sep=',', encoding='utf-8')
df_resumen.to_csv("resumen_tiendas_ntp.csv", sep=',', encoding='utf-8')
```

### Evoluci贸n de la tienda en el periodo


```python

lista_tiendas=df_ntp['Tienda'].unique().tolist()

# Documento en pdf
pdf_pages = PdfPages('reporte_ntp.pdf')

#Genera los graficos y los pone en el PDF
for tienda in lista_tiendas:
    fig, ax = plt.subplots(figsize=(15,7))
    df_ntp.loc[df_ntp['Tienda'] == tienda].groupby(['entrega','Tienda']).mean()['ntp'].unstack().plot(kind='bar',ax=ax)
    pdf_pages.savefig()
 
# Escribe el PDF en el archivo
pdf_pages.close()
```

# Procesamiento de KPIs de tiempo

```python
df_tiempo_entrega = pd.read_csv('clean_TIEMPO_DE_CONFIRMACION.csv',parse_dates=['Fecha'], date_parser=dateparse)
df_tiempo_entrega['Tiempo_confirmacion'] = df_tiempo_entrega['Tiempo_confirmacion'].convert_objects(convert_numeric=True)
df_tiempo_entrega

pysql = lambda q: pdsql.sqldf(q, globals())
strsql = "select * from df_tiempo_entrega group by tienda, Fecha order by tienda ASC"
df_tiempo_respuesta = pysql(strsql)
```

```python
lista_tiendas=df_tiempo_respuesta['Tienda'].unique().tolist()

pdf_pages = PdfPages('reporte_tiempo_respuesta.pdf')

for tienda in lista_tiendas:
    fig, ax = plt.subplots(figsize=(15,7))
    df_tiempo_respuesta.loc[df_tiempo_respuesta['Tienda'] == tienda].groupby(['Fecha','Tienda']).mean()['Tiempo_confirmacion'].unstack().plot(kind='bar',ax=ax)
    pdf_pages.savefig()
 
pdf_pages.close()
```

```python
p =  df_tiempo_respuesta.pivot(index='Fecha', columns='Tienda', values='Tiempo_confirmacion')
p.fillna(0, inplace=True)
df_resumen = p.describe()
p.to_csv("consolidado_tiendas_tiempo_respuesta.csv", sep=',', encoding='utf-8')
df_resumen.to_csv("resumen_tiendas_tiempo_respuesta.csv", sep=',', encoding='utf-8')
```
